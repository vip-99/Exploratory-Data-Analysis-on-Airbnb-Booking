# Exploratory Data Analysis (EDA)

 **1 How does the price of Airbnb listings vary by location (city or neighborhood)**

```python
price_variation=df.groupby(['neighbourhood_group'], as_index=False)['price'].var().sort_values(by='price', ascending=False)
price_variation
```

```python


# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['neighbourhood_group'], df['price'], color='yellow')

# Customize axis labels
plt.title('Neighbourhood of Variance by Location',weight='bold', color='purple')
plt.xlabel('Location')
plt.ylabel('Price')

# Add a subtle border
plt.gca().spines['top'].set_color('#d3d3d3')   # Light gray border for the top
plt.gca().spines['right'].set_color('#d3d3d3') # Light gray border for the right
plt.gca().spines['left'].set_linewidth(1.5)    # Slightly thicker left border
plt.gca().spines['bottom'].set_linewidth(1.5)  # Slightly thicker bottom border

# Show the plot with customizations
plt.show()
```

**Insight**

**Queens** exhibits the **highest** variation in **price** listings, while **Bronx** shows the **lowest**.

**14 Visualizing the correlation between various numerical features**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for Airbnb Data',fontsize=14,weight='bold', color='violet')
```

**Insight**


The correlation analysis reveals key insights:
**Hosts** with more listings have distinct host **IDs** (correlation = 0.58).
More listings are linked to slightly **higher prices** (correlation = 0.17).
**Hosts** with multiple listings have greater availability (correlation = 0.23).
Listings with more reviews tend to receive reviews more frequently (correlation = 0.58).

**Darker squares highlight stronger relationships, aiding in identifying factors influencing prices or other listing attributes.**
